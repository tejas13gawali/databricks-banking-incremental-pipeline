```python
# Load dependencies
from pyspark.sql.functions import *
from pyspark.sql.types import *
from delta.tables import DeltaTable

# Config Variables (Same as in 00_config)
catalog = "banking_cat"
schema  = "banking_db"

raw_path        = f"/Volumes/{catalog}/{schema}/raw"
bronze_path     = f"/Volumes/{catalog}/{schema}/bronze"
checkpoint_path = f"/Volumes/{catalog}/{schema}/checkpoint"

bronze_table = f"{catalog}.{schema}.bronze_transactions"
```

2. Create Bronze Table if not exists

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {bronze_table} (
    transaction_id STRING,
    account_id STRING,
    txn_date TIMESTAMP,
    amount DOUBLE,
    currency STRING,
    txn_type STRING,
    status STRING,
    created_at TIMESTAMP,
    source_file STRING,
    ingest_time TIMESTAMP
)
USING DELTA
LOCATION '{bronze_path}'
""")

print("Bronze table is ready:", bronze_table)
```
