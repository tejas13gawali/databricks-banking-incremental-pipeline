Create Catalog and Schema

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *

print("Starting configuration...")
```

Create Catalog and Schema

```python
spark.sql("CREATE CATALOG IF NOT EXISTS banking_cat")
spark.sql("CREATE SCHEMA IF NOT EXISTS banking_cat.banking_db")

result_catalog = spark.sql("SHOW CATALOGS LIKE 'banking_cat'")
result_schema = spark.sql("SHOW SCHEMAS IN banking_cat LIKE 'banking_db'")

# display(result_catalog)
# display(result_schema)
```

Create a Volume

```python
spark.sql("CREATE VOLUME IF NOT EXISTS banking_cat.banking_db.raw")
spark.sql("CREATE VOLUME IF NOT EXISTS banking_cat.banking_db.bronze")
spark.sql("CREATE VOLUME IF NOT EXISTS banking_cat.banking_db.silver")
spark.sql("CREATE VOLUME IF NOT EXISTS banking_cat.banking_db.gold")
spark.sql("CREATE VOLUME IF NOT EXISTS banking_cat.banking_db.checkpoint")

result_volumes = spark.sql("SHOW VOLUMES IN banking_cat.banking_db")
# display(result_volumes)

# print("Volumes are created and verified")
```

Define paths using volumes

```python
# Define paths using volumes

catalog = "banking_cat"
schema = "banking_db"

raw_path        = f"/Volumes/{catalog}/{schema}/raw"
bronze_path     = f"/Volumes/{catalog}/{schema}/bronze"
silver_path     = f"/Volumes/{catalog}/{schema}/silver"
gold_path       = f"/Volumes/{catalog}/{schema}/gold"
checkpoint_path = f"/Volumes/{catalog}/{schema}/checkpoint"

print("Raw path:", raw_path)
print("Bronze path:", bronze_path)
print("Silver path:", silver_path)
print("Gold path:", gold_path)
print("Checkpoint path:", checkpoint_path)
```

Define schema for banking transactions

```python
transaction_schema = StructType([
    StructField("transaction_id", StringType(), True),
    StructField("account_id", StringType(), True),
    StructField("txn_date", TimestampType(), True),
    StructField("amount", DoubleType(), True),
    StructField("currency", StringType(), True),
    StructField("txn_type", StringType(), True),
    StructField("status", StringType(), True),
    StructField("created_at", TimestampType(), True)
])
print("Schema ready.")
```
